```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For mapping and geocoding
import folium
from folium.plugins import MarkerCluster
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('data/Copy of Held For Sale 11.25.2025 1.xlsx', header=8)

df
```

```python
# Let's see all columns in the dataset
print("Columns in the dataset:")
print(df.columns.tolist())
print(f"\nShape: {df.shape}")
print(f"\nFirst few rows sample:")
df.head()
```

```python
# Check for existing coordinate columns
coord_columns = [col for col in df.columns if any(x in col.lower() for x in ['lat', 'lon', 'lng', 'cord', 'geo'])]
print("Potential coordinate columns found:", coord_columns if coord_columns else "None")

# Display all column names to understand the data structure
print("\nAll columns:")
for i, col in enumerate(df.columns):
    print(f"  {i}: {col}")
```

```python
# Initialize the geocoder
geolocator = Nominatim(user_agent="property_map_geocoder", timeout=10)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

def get_coordinates(row):
    """
    Try to get coordinates for a property using description and city.
    Returns (latitude, longitude) tuple or (None, None) if not found.
    """
    # Build search query from available data
    description = str(row.get('Description and Location of the Property', '')).strip()
    city = str(row.get('City', '')).strip()
    
    # Try different search strategies
    search_queries = []
    
    # Strategy 1: Full address with city and Puerto Rico
    if description and city:
        search_queries.append(f"{description}, {city}, Puerto Rico")
    
    # Strategy 2: Just city and Puerto Rico
    if city:
        search_queries.append(f"{city}, Puerto Rico")
    
    for query in search_queries:
        try:
            location = geocode(query)
            if location:
                return (location.latitude, location.longitude)
        except Exception as e:
            continue
    
    return (None, None)

# Check if lat/long columns already exist
lat_col = None
lon_col = None

for col in df.columns:
    col_lower = col.lower()
    if 'lat' in col_lower:
        lat_col = col
    if 'lon' in col_lower or 'lng' in col_lower:
        lon_col = col

# If coordinates exist, use them; otherwise geocode
if lat_col and lon_col:
    print(f"Using existing coordinate columns: {lat_col}, {lon_col}")
    df['Latitude'] = pd.to_numeric(df[lat_col], errors='coerce')
    df['Longitude'] = pd.to_numeric(df[lon_col], errors='coerce')
else:
    print("No coordinate columns found. Geocoding addresses...")
    print("This may take a few minutes due to rate limiting...")
    
    # Geocode each property
    coords = []
    total = len(df)
    
    for idx, row in df.iterrows():
        lat, lon = get_coordinates(row)
        coords.append((lat, lon))
        
        # Progress indicator
        if (idx + 1) % 10 == 0 or (idx + 1) == total:
            print(f"Processed {idx + 1}/{total} properties...")
    
    df['Latitude'] = [c[0] for c in coords]
    df['Longitude'] = [c[1] for c in coords]

# Report on geocoding success
geocoded = df[df['Latitude'].notna() & df['Longitude'].notna()]
print(f"\nSuccessfully geocoded: {len(geocoded)}/{len(df)} properties ({len(geocoded)/len(df)*100:.1f}%)")
```

```python
# Create the interactive map
def create_property_map(df):
    """Create an interactive folium map with property markers"""
    
    # Filter to properties with valid coordinates
    map_df = df[df['Latitude'].notna() & df['Longitude'].notna()].copy()
    
    if len(map_df) == 0:
        print("No properties with valid coordinates to display!")
        return None
    
    # Calculate center of map (Puerto Rico default if needed)
    center_lat = map_df['Latitude'].mean()
    center_lon = map_df['Longitude'].mean()
    
    # Default to Puerto Rico if coordinates seem off
    if pd.isna(center_lat) or pd.isna(center_lon):
        center_lat, center_lon = 18.2208, -66.5901  # Puerto Rico center
    
    # Create the map
    property_map = folium.Map(
        location=[center_lat, center_lon],
        zoom_start=9,
        tiles='OpenStreetMap'
    )
    
    # Add marker cluster for better performance with many markers
    marker_cluster = MarkerCluster().add_to(property_map)
    
    # Identify available columns for popup info
    all_columns = df.columns.tolist()
    
    # Common column names to display (adjust based on actual data)
    display_columns = []
    for col in all_columns:
        col_lower = col.lower()
        # Skip coordinate columns in popup
        if 'lat' in col_lower or 'lon' in col_lower or 'lng' in col_lower:
            continue
        display_columns.append(col)
    
    # Add markers for each property
    for idx, row in map_df.iterrows():
        # Build popup HTML content
        popup_html = '<div style="width:300px; max-height:400px; overflow-y:auto;">'
        popup_html += '<h4 style="margin-bottom:10px; color:#2c3e50; border-bottom:2px solid #3498db; padding-bottom:5px;">Property Details</h4>'
        popup_html += '<table style="width:100%; border-collapse:collapse;">'
        
        for col in display_columns:
            value = row.get(col, 'N/A')
            if pd.notna(value) and str(value).strip():
                # Format numbers nicely
                if isinstance(value, (int, float)):
                    if 'price' in col.lower() or 'value' in col.lower() or 'cost' in col.lower():
                        value = f"${value:,.2f}"
                    elif 'size' in col.lower() or 'sqft' in col.lower() or 'sq ft' in col.lower():
                        value = f"{value:,.0f} sq ft"
                    else:
                        value = f"{value:,.2f}" if isinstance(value, float) else f"{value:,}"
                
                popup_html += f'''
                <tr style="border-bottom:1px solid #eee;">
                    <td style="padding:5px; font-weight:bold; color:#34495e; width:40%;">{col}</td>
                    <td style="padding:5px; color:#2c3e50;">{value}</td>
                </tr>'''
        
        popup_html += '</table></div>'
        
        # Create popup
        popup = folium.Popup(popup_html, max_width=350)
        
        # Get property name/description for tooltip
        tooltip_text = str(row.get('Description and Location of the Property', 
                         row.get('REO #', f'Property {idx}')))[:50]
        
        # Add marker
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=popup,
            tooltip=tooltip_text,
            icon=folium.Icon(color='blue', icon='home', prefix='fa')
        ).add_to(marker_cluster)
    
    print(f"Map created with {len(map_df)} property markers")
    return property_map

# Create and display the map
property_map = create_property_map(df)
property_map
```

```python
# Save the map to an HTML file for easy sharing/viewing
if property_map:
    property_map.save('property_map.html')
    print("Map saved to 'property_map.html' - open this file in a browser for the full interactive experience!")
    print("\nYou can:")
    print("  - Click on markers to see property details")
    print("  - Zoom in/out and pan around the map")
    print("  - Click on clusters to expand them")
```

```python
# Show properties that couldn't be geocoded (if any)
not_geocoded = df[df['Latitude'].isna() | df['Longitude'].isna()]

if len(not_geocoded) > 0:
    print(f"\n⚠️ {len(not_geocoded)} properties could not be geocoded:")
    display_cols = ['REO #', 'Description and Location of the Property', 'City']
    display_cols = [c for c in display_cols if c in df.columns]
    print(not_geocoded[display_cols].to_string())
else:
    print("✅ All properties were successfully geocoded!")
```
